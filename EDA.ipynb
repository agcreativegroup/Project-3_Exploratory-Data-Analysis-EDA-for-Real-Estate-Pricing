# Handling missing value

```python
print("\nMissing Values Before Handling:")
print(housing_data.isnull().sum()[housing_data.isnull().sum() > 0])  # Show only columns with missing values
```

```python
# Impute missing numerical values with median
if 'LotFrontage' in housing_data.columns:
    housing_data.loc[:, 'LotFrontage'] = housing_data['LotFrontage'].fillna(housing_data['LotFrontage'].median())
```

```python
# Impute categorical missing values with mode or a category like 'None'
if 'Alley' in housing_data.columns:
    housing_data.loc[:, 'Alley'] = housing_data['Alley'].fillna('None')
```

```python
# Instead of dropping, replace missing categorical values
if 'GarageType' in housing_data.columns:
   housing_data.loc[:,'GarageType']= housing_data['GarageType'].fillna('None')
```

```python
# Removing Duplicates
housing_data.drop_duplicates(inplace=True)
```

```python
housing_data
```

```python
print("\nMissing Values After Handling:")
print(housing_data.isnull().sum()[housing_data.isnull().sum() > 0])
```

```python
# Addressing Outliers Using Z-Score 
numeric_cols = housing_data.select_dtypes(include=[np.number]).columns
z_scores = np.abs((housing_data[numeric_cols] - housing_data[numeric_cols].mean()) / housing_data[numeric_cols].std())
housing_data = housing_data[(z_scores < 3).all(axis=1)]  #  values within 3 standard deviations
```

```python
# Removing Rows with Missing Values
data_cleaned = housing_data.dropna()  # Removes rows where any column has NaN
```

```python
# Removing Columns with Too Many Missing Values
data_cleaned = housing_data.dropna(axis=1, thresh=int(0.5 * len(housing_data)))  # Drops columns with >50% missing data
```

```python
# Removing Duplicates Based on Specific Columns
if 'GrLivArea' in housing_data.columns and 'Electrical' in housing_data.columns:
    data_cleaned = housing_data.drop_duplicates(subset=['GrLivArea', 'Electrical'])

print("\nData After Cleaning:", data_cleaned.head())
```

```python
# Correcting Inconsistent Categorical Values 
if 'HeatingQC' in housing_data.columns:
    housing_data.loc[:, 'HeatingQC'] = housing_data['HeatingQC'].replace({'incorrect_value': 'correct_value'})

```

```python
# Convert column to numeric 
housing_data.loc[:, "1stFlrSF"] = pd.to_numeric(housing_data["1stFlrSF"], errors="coerce")

#  log transformation 
housing_data.loc[:, "1stFlrSF"] = np.log1p(housing_data["1stFlrSF"])
```

```python
housing_data = housing_data.drop(columns=['Unnamed: 0'])
```

```python
print("\nFinal Cleaned Data Overview:")
print(housing_data.info())
```
