```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  visual settings
sns.set(style="whitegrid")
```

```python
# Load the dataset 
housing_data = pd.read_csv("housing_data.csv")

# Select relevant numeric columns for analysis
numeric_columns = ['SalePrice', 'LotArea', 'GrLivArea', 'TotalBsmtSF', '1stFlrSF', 'OverallQual', 'YearBuilt']

# Subset the dataset for multivariate analysis
housing_subset = housing_data[numeric_columns]

# Handle missing values by filling with median 
housing_subset.loc[:, :] = housing_subset.fillna(housing_subset.median())

```

# Correlation Matrix and Heatmap

```python
# Compute the correlation matrix
corr_matrix = housing_subset.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Key Features")
plt.show()
```

# Pairplot (Scatterplot Matrix)

```python
# Scatterplot matrix to visualize pairwise relationships
sns.pairplot(housing_subset, diag_kind='kde', markers='+', corner=True)
plt.show()
```

# Individual Scatterplots for Key Relationships

# Sale Price vs. Living Area

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=housing_subset['GrLivArea'], y=housing_subset['SalePrice'], alpha=0.5, color='blue')
plt.title("Sale Price vs. Living Area")
plt.xlabel("Above Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.grid(True)
plt.show()
```

# Sale Price vs. Overall Quality

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='OverallQual', y='SalePrice', data=housing_subset, palette="viridis", hue='OverallQual', legend=False)

plt.title("Sale Price vs. Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("Sale Price")
plt.grid(True)
plt.show()
```

# Regression Plot 

```python
plt.figure(figsize=(10, 6))

# Regression plot: Sale Price vs. Total Basement SF
sns.regplot(x=housing_data['TotalBsmtSF'], y=housing_data['SalePrice'], scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.xlabel('Total Basement SF')
plt.ylabel('Sale Price')
plt.title('Sale Price vs. Total Basement SF')
plt.show()
```
